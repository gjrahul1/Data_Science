### Let's find the cosine complexity between:

<b> 1. </b> [[3,1]
 [6,2
]]

<b> 2. </b> [[
3,0]
[0,8
]]

<b> 3. </b> [[
3,1]
[3,2
]]

```python
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
```

```python
#Similartiy 1
cosine_similarity([[3,1]],[[6,2]])
```

- From the cosine similarity, we can conclude that the data points [3,1] & [6,2] are parallel to each other

```python
#Similarity 2
cosine_similarity([[3,0]],[[0,8]])
```

- From the cosine similarity, we can conclude that the data points [3,0] & [0,8] is 90 degree to each other

```python
#Similarity 3
cosine_similarity([[3,1]],[[3,2]])
```

- From the cosine similarity, we can conclude that the data points [3,1] & [3,2] is 0.964 which suggests that the points are closer to each other

### Visualization
- Visualizing the data points

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Define the vectors
vector1 = np.array([3,1])
vector2 = np.array([6,2])
vector3 = np.array([3,0])
vector4 = np.array([0,8])
vector5 = np.array([3,1])
vector6 = np.array([3,2])
```

```python
#Calculate endpoints for plotting
origin = np.array([0,0])
endpoint1 = origin + vector1
endpoint2 = origin + vector2
endpoint3 = origin + vector3
endpoint4 = origin + vector4
endpoint5 = origin + vector5
endpoint6 = origin + vector6
```

```python
#Setup the plot
sns.set(style='whitegrid') 
plt.figure(figsize=(8,8))
```

```python
#Plot the vectors
plt.quiver(*origin,*vector1,scale=1, scale_units='xy',angles = 'xy', color = 'r', label='Vector 1')
plt.quiver(*origin, *vector2, scale=1, scale_units='xy',angles='xy', color='b', label='Vector 2')
plt.quiver(*origin, *vector3, scale = 1, scale_units='xy',angles = 'xy', color = 'c', label='Vector 3')
plt.quiver(*origin, *vector4, scale = 1,scale_units='xy',angles = 'xy', color = 'm', label = 'Vector 4')
plt.quiver(*origin, *vector5, scale = 1, scale_units = 'xy',angles = 'xy', color = 'k', label = 'Vector 5')
plt.quiver(*origin, *vector6, scale = 1, scale_units = 'xy',angles = 'xy',color = '0.8', label = 'Vector 6')

#Set the plot limits
plt.xlim(-1,7)
plt.ylim(-1,8)

#Add labels and titles
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Comparsion between cosine-similarity')
plt.legend()

plt.show()
```
