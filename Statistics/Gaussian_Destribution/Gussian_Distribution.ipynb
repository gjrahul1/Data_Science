```python
from numpy import arange
from matplotlib import pyplot as plt
from scipy.stats import norm
```

#### Probability Density Distribution
##### pdf(x,loc=0,scale=1)

<b> x:</b> means data to be passed

<b> loc: </b> means mean

<b> scale: </b> means standard deviation


```python
#X-axis for the plot
x_axis = arange(-3,3,0.001)

#y-axis as gaussian
y_axis = norm.pdf(x_axis,0,1)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

##### Data Sample and Data Distribution

Data Sample: A subset of observations from a group

Data Population: All possible observations from a group

```python
import numpy as np
```

```python
np.random.seed(1)
data = 5 * np.random.randn(10000) + 50
```

```python
#Let's see what data contains for us
begin = 0
for value in data:
    begin += 1
    print(value)
    print(f'{begin} of {len(data)}')
```

```python
data_min = data.min()
data_max = data.max()

print(f"Minimum: {data_min}")
print(f'Maximum: {data_max}')
```

#### Observation

- We can understand that we are having random values between 31-67
- We have 10,000 such values
