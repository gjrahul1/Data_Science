#### Data Normalization
- Let's normalize the revenue data
- To normalize the data, I would use divide each data point using 10^6 since each data point is in 10^6

```python
#Normalizing the revenue data
df['normalized_revenue'] = df['revenue'].apply(lambda x: x/1000000)
```

```python
#Descrbing the data
df['normalized_revenue'].describe()
```

```python
_,mean,std,*_ = df['normalized_revenue'].describe()
print(f"mean:{mean} \nstandard deviation: {std}")
```

```python
#Function for z score
def get_z_score(value,mean,std):
    return (value-mean)/std

df['zscore'] = df.normalized_revenue.apply(lambda x: get_z_score(x,mean,std))
```

```python
print(df['zscore'].head(3))
```

```python
#Function for Mean Absolute Distribution
def get_MAD(s):
    median = np.median(s)
    diff = abs(s - median)
    MAD = np.median(diff)
    return MAD
```

```python
#Defining variables
MAD = get_MAD(df.normalized_revenue)
median = np.median(df.normalized_revenue)

#Modified zscore
def get_modified_zscore(value,median,MAD):
    return 0.675*(value-median)/MAD
```

```python
#Calculating Modified Z-Score
df['mod_zscore'] = df.normalized_revenue.apply(lambda x: get_modified_zscore(x,median,MAD))
```
