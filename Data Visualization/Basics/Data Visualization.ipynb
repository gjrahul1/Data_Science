### Data Visualization
This notebook aims to cover data visualization in brief

```python
import numpy as np
from random import seed, randint
from matplotlib import pyplot as plt
%matplotlib inline
```

#### Line Chart

```python
x = [x*0.1 for x in range(100)]
```

```python
#Print first few sample of our data
def print_values(x):
    limit = 5
    if limit > 0:
        for data in x:
            print("{:.2f}".format(data))
            limit -=1
            if limit == 0:
                break
```

```python
print_values(x)
```

##### Observation 
- We can see we have values from 0.0, 0., 0.2 until 10.0

```python
#Y Axis
y = np.sin(x)
```

```python
print_values(y)
```

##### Observation 
- We can see we have values from 0.0, 0., 0.2

```python
#Plotting the graph
plt.plot(x,y)
```

#### Bar Chart

```python
#Seed the random number generator
seed(1)

#names for the categories
x = ['red','green','blue']

#Quantities for the each category
y = [randint(0,100), randint(0,100),randint(0,100)]
```

```python
#Bar chart
plt.xlabel('Categories')
plt.ylabel('Quantities')
plt.bar(x,y,color=['red','green','blue'])
```

#### Histogram Plot

```python
#Random Seed Generator
seed(1)
#Random number drawn from Gaussian Distribution
x = np.random.randn(1000)
```

```python
res = [value for value in x]
# print(res)
```

```python
#Plotting histogram
plt.hist(x,bins=100)
plt.show()
```

#### Box and Whisker Plot

- A box and whisker plot for short, generally used to summarize the distribution  of data sample
- The boxplot is a graphical representation of the <b>Five Number Summary</b>
- X axis is used to represent the samples
- Y axis is used to represent the observations

```python
#Seed the random number generator
seed(1)

#Random numbers drawn from gaussian distribution
x = [np.random.randn(1000), 5 * np.random.randn(1000), 10 * np.random.randn(1000)]
```

```python
plt.boxplot(x)
plt.show()
```

### Scatter Plot
- Scatter plots are bivariate or trivariate plots of variables against each other

```python
#X-Axis
x = 20 * np.random.randn(1000) + 100

#Y-Axis
y = x + (10 * np.random.randn(1000) + 50)
```

```python
plt.scatter(x,y)
plt.show()
```

- From the above graph, we can suggest that the relationship between the two variables are positive
