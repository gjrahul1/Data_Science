#### Law of Large Numbers

```python
#Idealized Population Distribution
from numpy import arange
from matplotlib import pyplot as plt
from scipy.stats import norm
```

```python
x_axis = arange(30,71,1)
y_axis = norm.pdf(x_axis,50,5)
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

##### Calculate error in the population mean vs sample size

```python
from numpy import mean,array
from numpy.random import randn,seed
from matplotlib import pyplot as plt
```

```python
#Seed
seed(1)
```

```python
#Sample Sizes
sizes = list()
```

```python
for x in range(10,20000,200):
    sizes.append(x)
```

```python
#Calculate mean
means = [mean(5 * randn(size) + 50) for size in sizes]
```

```python
plt.scatter(sizes,array(means)-50)
plt.show()
```
